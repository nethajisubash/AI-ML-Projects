```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def J(x) :
  return (x*x)-(4*x) +3

def dJ(x) :
  return (2*x)-4

x=np.linspace(-20,20,1000)
y=J(x)
plt.plot(x,y)
plt.show()
```

```python
alpha = random.uniform(0,1)
iterations=100
x_init=random.randint(0,10000)
print("INitial x  {} : Learning rate  {} ".format(x_init,alpha))


def gradient_descent(J,x,alpha,iterations) :
  J_history = []
  for i in range(iterations) :
    x = x - alpha * dJ(x)
    J_history.append(J(x))
  print ("x Optimized {} : cost Min {}".format(x,J(x)))
  return x, J_history

x,j_hist = gradient_descent(J,x_init,alpha, iterations)


```
